```python
def team_weights(name):
  name = name
  team_df = raw_df.loc[((raw_df['team1'] == name) | (raw_df['team2'] == name))]
  team_df.dropna()
  team_df.drop(columns =['margin','man_of_the_match','winner','place'], axis=1, inplace=True)
  team_df.reset_index(inplace = True)
  len(team_df['team1'].values)
  team_score = []
  for i in range(len(team_df)):
    if team_df['team1'].values[i] == 'name':
      team_score.append(team_df['team1_score'].values[i])
    else:
      team_score.append(team_df['team2_score'].values[i])
  team_df.drop(columns =['team1_score','team2_score','index'], axis=1, inplace=True)
  for i in range(len(team_df)):
    if team_df['toss_winner'].values[i] == 'name':
      team_df['toss_winner'].values[i] = 1
    else:
      team_df['toss_winner'].values[i] = 0
  opponent_team = []
  for i in range(len(team_df)):
    if team_df['team1'].values[i] != 'name':
      opponent_team.append(team_df['team1'].values[i])
    else:
      opponent_team.append(team_df['team2'].values[i])
  team_df['Opponent_Team'] = opponent_team
  team_df.drop(columns =['team1','team2'], axis=1, inplace=True)
  team_df['team score'] = team_score
  columns_to_encode = ['stadium', 'Opponent_Team']
  team_OneHot_df = pd.get_dummies(team_df, columns=columns_to_encode, drop_first=True)
  team_OneHot_df. drop(columns=['year '], inplace=True)
  team_OneHot_df.insert(1, 'toss_choice_Bat', '')
  for i in range(len(team_df)):
    if team_df['toss_choice'].values[i] == 'Bat':
      team_OneHot_df['toss_choice_Bat'].values[i] = 1
    else:
      team_OneHot_df['toss_choice_Bat'].values[i] = 0
  x = team_OneHot_df.drop('team score', axis=1)
  x. drop(columns=['toss_choice'], inplace=True)
  column_names = x.columns.tolist()
  print(column_names)
  y = team_OneHot_df['team score']
  X = np.array(x)
  #print(X)
  Y = np.array(y)
  #print(Y)
  theta = np.random.rand(len(X[0]))
  H_Theta = np.dot(X,theta)
  SUM = []
  for i in range(len(X[0])):
      SUM.append(0)
  for i in range(len(X)):
    SUM+=((Y[i]-H_Theta[i])*X[i])
  for i in range(1):
    alpha = 0.01
    THETA = []
    COST = []
    for i in range(5):
        theta = theta + alpha*SUM
        THETA.append(theta)
        H_Theta = np.dot(X,theta)
        cost = 0
        for j in range(len(X)):
          SUM = SUM + (Y[j] - H_Theta[j])*X[j]/len(X)
          cost+=(H_Theta[j] - Y[j])**2/len(X)
        COST.append(cost)
        #if (i%2000)== 0:
            #print(cost, i)
  min_cost = np.min(COST)
  min_index = np.argmin(COST)
  theta = THETA[min_index]
  return theta, x
  #print("Minimum value:", min_cost)
  #print("Index of minimum value:", min_index)
  #print(x)

```
